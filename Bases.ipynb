```python
def negabase(i, b):
    if b < 0:
        digits = []
        if not i:
            digits = ['0']
        else:
            while i != 0:
                i, remainder = divmod(i, b)
                if remainder < 0:
                    i, remainder = i + 1, remainder -b
                digits.append(str(remainder))
            result = ''.join(digits[::-1])
    else: 
        result = 'Please enter a negative number for the base'
    return result
```

```python
def bases(i, b):
    
    pos_digits = []
    neg_digits = []
    i_neg = i
    i_pos = i
    alpha = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
    if not i:
        neg_digits = ['0']
        pos_digits = ['0']
    else:
        while i_neg != 0:
            i_neg, nremainder = divmod(i_neg, -1*b)
            if nremainder < 0:
                i_neg, nremainder = i_neg + 1, nremainder + b
            if nremainder > 9:
                nrem_str = alpha[nremainder - 10]
            else:
                nrem_str = str(nremainder)
            neg_digits.append(nrem_str)
        neg_base = ''.join(neg_digits[::-1])
        
        while i_pos != 0:
            i_pos, premainder = divmod(i_pos, b)
#            if premainder < 0:
#                i_pos, premainder = i_pos + 1, premainder + b
            if premainder > 9:
                prem_str = alpha[premainder - 10]
            else:
                prem_str = str(premainder)
            pos_digits.append(prem_str)
        pos_base = ''.join(pos_digits[::-1])
        
    return pos_base, neg_base
```

```python
bases(31, 16)
```

```python
for bn in range(2,10):
        print(bn, bases(13,bn))
```

```python
for bn in range(2,36):
    for j in range(1,1000):
        p, n = bases(j,bn)
        if p == n:
            print(j,' in base ',bn,bases(j,bn))
```
